# 計算量のグラフを書いてみる

計算量の実態に応じたグラフを描いてみる

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def polynomial(n, c = 1):
    '''多項式オーダー n の c 乗'''
    return n**c

def logarithmic(n):
    '''対数オーダー log(n)'''
    return np.log(n)

def constant(n):
    '''定数オーダー'''
    return np.ones(n.shape)

def exponential(n):
    '''指数オーダー 2^n'''
    return 2**n

```

まずは $n \leq 50$ 程度で，定数オーダー $\cal{O}(1)$，
線形オーダー $\cal{O}(n)$，
対数オーダー $\cal{O}(\log n)$
の増加を比べてみよう

```python
n = np.arange(1, 51)

o_1 = constant(n)
o_logn = logarithmic(n)
o_n = polynomial(n, 1)

plt.plot(n, o_1, label='O(1)')
plt.plot(n, o_logn, label='O(log n)')
plt.plot(n, o_n, label='O(n)')
plt.grid()
plt.legend()
```

定数オーダー $\cal{O}(1)$，対数オーダー $\cal{O}(\log n)$，線形オーダー $\cal{O}(n)$ を比較してみると，
$\cal{O}(\log n)$ は，実際のところかなり増加が少ないのがわかる．

次に，線形オーダー $\cal{O}(n)$と，多項式オーダー $\cal{O}(n^2)$，指数オーダー $\cal{O}(2^{0.1n})$ を比べてみる

```python
n = np.arange(1, 51)

o_n = polynomial(n, 1)
o_n2 = polynomial(n, 2)
o_2n = exponential(0.1* n)

plt.plot(n, o_n, label='O(n)')
plt.plot(n, o_n2, label='O(n^2)')
plt.plot(n, o_2n, label='O(2^(0.1*n))')
plt.grid()
plt.legend()
```

べきオーダー $\cal{O}(2^{0.1 n})$ はそれほどではないようにみえるが，$n$ を 200 程度までに増やしてみる

```python
# n を増やす
n = np.arange(1, 201)

o_n = polynomial(n, 1)
o_n2 = polynomial(n, 2)
o_2n = exponential(0.1* n)

plt.plot(n, o_n, label='O(n)')
plt.plot(n, o_n2, label='O(n^2)')
plt.plot(n, o_2n, label='O(2^(0.1*n))')
plt.grid()
plt.legend()
```

$n$ が 150 を超えたあたりで，増加率の違いは明白になる．．．なので十分 $n$ が大きいケースでのべきオーダーの計算はかなり問題になることがわかる．

```python
# 縦軸を対数表示で描いてみる

n = np.arange(1, 51)
o_1 = constant(n)
o_logn = logarithmic(n)
o_n = polynomial(n, 1)
o_n2 = polynomial(n, 2)
o_2n = exponential(n)

plt.semilogy(n, o_1, label='O(1)')
plt.semilogy(n, o_logn, label='O(log n)')
plt.semilogy(n, o_n, label='O(n)')
plt.semilogy(n, o_n2, label='O(n^2)')
plt.semilogy(n, o_2n, label='O(2^n)')
plt.grid()
plt.legend()
```
